13. Now we want to know which part of the week the most last updates happened. For this you need to group `Last Updated` column into `weekday` and `weekend`, then visualize the result as a `barplot`. You should use a vizualization package.

```python
# your code here
google['Last Updated']= pd.to_datetime(google['Last Updated'],format='%B %d, %Y',errors='coerce')
```

```python
# Creating 2 df which are weekend and weekday
lnd = []
ld = []

for i in google['Last Updated'].dt.day_name():
    if i == 'Monday' or i == 'Tuesday' or i == 'Wednesday' or i == 'Thursday' or i == 'Friday':
        ld.append(i)

for k in google['Last Updated'].dt.day_name():  
    if k == 'Saturday' or k == 'Sunday':
        lnd.append(k)

weekday = pd.DataFrame(ld, columns=["Weekday"])
weekend = pd.DataFrame(lnd, columns=["Weekend"])


```

```python
import matplotlib.pyplot as plt

weekday_counts = weekday['Weekday'].value_counts()
weekend_counts = weekend['Weekend'].value_counts()


plt.bar(weekday_counts.index, weekday_counts.values, color='blue', label='Weekday')
plt.bar(weekend_counts.index, weekend_counts.values, color='green', label='Weekend')


plt.xlabel('Day of the Week')
plt.ylabel('Number of Last Updates')
plt.title('Number of Last Updates by Day Type')
plt.legend()


plt.show()
```
